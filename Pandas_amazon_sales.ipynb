### Cuántas categorías hay y cuántos productos hay por categoría?

```python
# Cómo vemos que estas categorías son muy largas y poco entendibles vamos a 
#subdividirlas por el contenido entre | 

# As we see that these categories are very long and not very understandable we are going to 
#subdivide them by the content between | 

df[['category_1', 'category_2', 'category_3', 'category_4']] = df['category'].str.split('|', expand=True, n=3)
df.head()
```

```python
# Cogemos la 3 porque creo que es de las más reprentativas (sin ver el ecommerce y viendo sólo el dataset)
# We take 3 because I think it is one of the most representative (without seeing the ecommerce and seeing only the dataset).

df["category_3"].nunique()

# Vemos que hay 71 categorías distintas (en category a secas hay 211)
# We see that there are 71 different categories (in category a secas there are 211).
```

```python
conteo_por_categoria = df["category_3"].value_counts()
```

```python
# Lo ponemos en formato df para que quede más ordenado
# We put it in df format to make it more orderly.
conteo_por_categoria_df = conteo_por_categoria.reset_index()
conteo_por_categoria_df.columns = ["category_3", "count"]

print("Número de productos distintos por categoría en forma de DataFrame:")
print(conteo_por_categoria_df)

# Vemos el número de productos que hay en cada categoría
# We see the number of products in each category
```

```python
# Si queremos buscar uno que no sale podríamos usar la función iloc, ej:
# If we want to search for one that does not come up, we could use the iloc function, e.g:
conteo_por_categoria_df.iloc[20]
```

#### Gráfico de las 5 categorías con más productos en matplotlib

```python
# Escogemos las 5 primeras:
# We choose the first 5:
top_5 = conteo_por_categoria_df.nlargest(5, "count")

# Ordenamos el df de mayor a menor porque prefiero ese orden para la visualización:
# We order the df from highest to lowest because I prefer that order for visualisation:
top_5 = top_5.sort_values("count", ascending = True)

# Creamos el gráfico de barras en horizontal
# We create the horizontal bar chart
plt.barh(top_5["category_3"], top_5["count"], color="#FF9900")
plt.xlabel("Number of products")
plt.ylabel("Categories")
plt.show()
```

```python
# Cómo vemos que los precios tienen un formato raro vamos a ponerlo sólo como numérico y luego en el gráfico asumir que son dólares
# As we see that the prices have a strange format we are going to put it only as numerical and then in the graph assume that they are dollars.
df['discounted_price'] = df['discounted_price'].str.replace('₹', '')
df['actual_price'] = df['actual_price'].str.replace('₹', '')
```

```python
# Buscamos los 5 productos más vendidos por el product_id
# We search for the top 5 products by product_id

products_count = df["product_id"].value_counts()
top_5_products = products_count.nlargest(5)
top_5_products_filtered_df = df[df["product_id"].isin(top_5_products.index)]
top_5_products_filtered_df.head()
```

```python
top_5_products
```

```python
# Creamos un DataFrame con dos columnas: product_id y sell_products
top_5_products = pd.DataFrame({"product_id": top_5_products.index, "sell_products": top_5_products.values})
```

```python
# Lo juntamos con el df original para obtener también el precio y el descuento
df_top_5_products = pd.merge(df[["product_id", "discounted_price", "discount_percentage", "actual_price"]], top_5_products, on="product_id")
df_top_5_products
```

#### Histograma de ratings
#### Histogram of ratings

```python
df.columns
```

```python
df.dtypes
```

```python
df = df[df["rating"] != "|"]

# Convierte la columna 'rating' a tipo float
df["rating"] = df["rating"].astype(float)
```

```python
# Definimos los valores de rating que deseas mostrar en el eje x
# We define the rating values you want to display on the x-axis.
rating_values = np.linspace(df["rating"].min(), df["rating"].max(), 10)


sns.histplot(df["rating"], bins=10, kde=True, color="skyblue")
plt.gca().xaxis.set_major_formatter(plt.matplotlib.ticker.StrMethodFormatter("{x:.2f}"))
plt.title("Distribución de Ratings")
plt.xlabel("Rating")
plt.ylabel("Número de Productos")
plt.show()
```

Podemos ver que la mayoría de productos están entre 4 y 4.5 estrellas, por ello en general estos productos tienen buenos ratings. Quizás habría que ver los productos menores a x rating.

We can see that most products are between 4 and 4.5 stars, so in general these products have good ratings. Perhaps the products below x rating should be looked at.
